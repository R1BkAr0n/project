```python
# @title 10. –†–∞—Å—á—ë—Ç –ª–∏–Ω–∑—ã (—Å —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ–º)
async def calculate_lens(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        # –†–∞–∑–±–∏–≤–∞–µ–º –≤–≤–µ–¥—ë–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç –Ω–∞ —á–∏—Å–ª–∞
        values = update.message.text.split()
        f_values = []

        for val in values:
            try:
                f = float(val.replace(',', '.'))
                if 2 <= f <= 100:  # –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Ä–∞–∑—É–º–Ω–æ—Å—Ç—å
                    f_values.append(f)
                else:
                    await update.message.reply_text(f'‚ö†Ô∏è –ó–Ω–∞—á–µ–Ω–∏–µ {f} –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ (2-100 —Å–º) - –ø—Ä–æ–ø—É—â–µ–Ω–æ')
            except:
                await update.message.reply_text(f'‚ö†Ô∏è "{val}" –Ω–µ —á–∏—Å–ª–æ - –ø—Ä–æ–ø—É—â–µ–Ω–æ')

        if not f_values:
            await update.message.reply_text('‚ùå –ù–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π!')
            return

        # –£—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ
        avg_f = sum(f_values) / len(f_values)

        # –†–∞—Å—á—ë—Ç
        F = avg_f
        D = 1 / (F / 100)

        # –ê–Ω–∞–ª–∏–∑
        response = "üîç **–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –ª–∏–Ω–∑—ã**\n\n"
        response += f"üìä –ò–∑–º–µ—Ä–µ–Ω–∏–π: {len(f_values)}\n"
        response += f"üìè –ó–Ω–∞—á–µ–Ω–∏—è f: {', '.join([str(f) for f in f_values])} —Å–º\n"
        response += f"üìà –°—Ä–µ–¥–Ω–µ–µ f = {avg_f:.1f} —Å–º\n"
        response += f"üîç F = {F:.1f} —Å–º\n"
        response += f"üëì D = {D:.1f} –¥–ø—Ç—Ä\n\n"

        response += "üìä **–ê–Ω–∞–ª–∏–∑ –æ—à–∏–±–æ–∫:**\n"

        # –†–∞–∑–±—Ä–æ—Å –∏–∑–º–µ—Ä–µ–Ω–∏–π
        if len(f_values) > 1:
            min_f = min(f_values)
            max_f = max(f_values)
            razbros = max_f - min_f
            response += f"‚Ä¢ –†–∞–∑–±—Ä–æ—Å –∑–Ω–∞—á–µ–Ω–∏–π: {razbros:.1f} —Å–º\n"

            if razbros < 1:
                response += "‚úÖ –ò–∑–º–µ—Ä–µ–Ω–∏—è –æ—á–µ–Ω—å —Ç–æ—á–Ω—ã–µ!\n"
            elif razbros < 3:
                response += "üëç –•–æ—Ä–æ—à–∞—è –ø–æ–≤—Ç–æ—Ä—è–µ–º–æ—Å—Ç—å\n"
            elif razbros < 5:
                response += "ü§î –°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å ‚Äì —Ñ–æ–∫—É—Å–∏—Ä—É–π—Å—è —Ç–æ—á–Ω–µ–µ\n"
            else:
                response += "‚ùå –ë–æ–ª—å—à–æ–π —Ä–∞–∑–±—Ä–æ—Å ‚Äì –ø—Ä–æ–≤–µ—Ä—å –º–µ—Ç–æ–¥–∏–∫—É\n"

        # –•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –ª–∏–Ω–∑—ã
        if D > 20:
            response += "‚Ä¢ –≠—Ç–æ –æ—á–µ–Ω—å —Å–∏–ª—å–Ω–∞—è –ª–∏–Ω–∑–∞ (–º–∏–∫—Ä–æ—Å–∫–æ–ø)\n"
        elif D > 10:
            response += "‚Ä¢ –≠—Ç–æ —Å–∏–ª—å–Ω–∞—è –ª–∏–Ω–∑–∞ (–ª—É–ø–∞)\n"
        elif D > 5:
            response += "‚Ä¢ –≠—Ç–æ —Å—Ä–µ–¥–Ω—è—è –ª–∏–Ω–∑–∞ (–æ—á–∫–∏)\n"
        else:
            response += "‚Ä¢ –≠—Ç–æ —Å–ª–∞–±–∞—è –ª–∏–Ω–∑–∞\n"

        if avg_f < 5:
            response += "‚ö†Ô∏è –û—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∏–π —Ñ–æ–∫—É—Å ‚Äì –≤–æ–∑–º–æ–∂–Ω–æ, –ª–∏–Ω–∑–∞ –Ω–µ —Ç–∞\n"
        elif avg_f > 30:
            response += "‚ö†Ô∏è –ë–æ–ª—å—à–æ–π —Ñ–æ–∫—É—Å ‚Äì –ø—Ä–æ–≤–µ—Ä—å, —Ç–æ—á–Ω–æ –ª–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ –±—ã–ª–æ —á—ë—Ç–∫–∏–º\n"

        await update.message.reply_text(response, parse_mode='Markdown')

    except Exception as e:
        await update.message.reply_text(f'‚ùå –û—à–∏–±–∫–∞: {str(e)}')
```

```python
# @title 11. –†–∞—Å—á—ë—Ç –º–∞—è—Ç–Ω–∏–∫–∞ (—Å —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ–º)
async def calculate_pendulum(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        parts = update.message.text.split()
        if len(parts) < 2:
            await update.message.reply_text('‚ùå –ù—É–∂–Ω–æ –º–∏–Ω–∏–º—É–º 2 —á–∏—Å–ª–∞: L –∏ t1 t2 ...')
            return

        L = float(parts[0].replace(',', '.'))

        # –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –≤—Ä–µ–º–µ–Ω–∞
        times = []
        for val in parts[1:]:
            try:
                t = float(val.replace(',', '.'))
                if t > 0:
                    times.append(t)
                else:
                    await update.message.reply_text(f'‚ö†Ô∏è –í—Ä–µ–º—è {t} <= 0 - –ø—Ä–æ–ø—É—â–µ–Ω–æ')
            except:
                await update.message.reply_text(f'‚ö†Ô∏è "{val}" –Ω–µ —á–∏—Å–ª–æ - –ø—Ä–æ–ø—É—â–µ–Ω–æ')

        if not times:
            await update.message.reply_text('‚ùå –ù–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏!')
            return

        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–ª–∏–Ω—ã
        if L <= 0:
            await update.message.reply_text('‚ùå –î–ª–∏–Ω–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π')
            return

        if L > 2:
            await update.message.reply_text('‚ö†Ô∏è L > 2 –º? –ü—Ä–æ–≤–µ—Ä—å, –º–æ–∂–µ—Ç –≤ —Å–∞–Ω—Ç–∏–º–µ—Ç—Ä–∞—Ö?')

        # –†–∞—Å—á—ë—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑–º–µ—Ä–µ–Ω–∏—è
        g_values = []
        results = []

        for t in times:
            if t < 10:
                await update.message.reply_text(f'‚ö†Ô∏è –í—Ä–µ–º—è {t} —Å —Å–ª–∏—à–∫–æ–º –º–∞–ª–æ –¥–ª—è 20 –∫–æ–ª–µ–±–∞–Ω–∏–π')
                continue

            T = t / 20
            g = 4 * math.pi**2 * L / (T**2)
            g_values.append(g)
            results.append((t, T, g))

        if not g_values:
            await update.message.reply_text('‚ùå –ù–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π –ø–æ—Å–ª–µ –ø—Ä–æ–≤–µ—Ä–æ–∫')
            return

        # –£—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ
        avg_g = sum(g_values) / len(g_values)
        error = abs(avg_g - G_TABLE) / G_TABLE * 100

        # –ê–Ω–∞–ª–∏–∑
        response = "‚è± **–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–∞—è—Ç–Ω–∏–∫–∞**\n\n"
        response += f"üìè L = {L} –º\n"
        response += f"üìä –ò–∑–º–µ—Ä–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏: {len(times)}\n\n"

        response += "**–û—Ç–¥–µ–ª—å–Ω—ã–µ –∏–∑–º–µ—Ä–µ–Ω–∏—è:**\n"
        for i, (t, T, g) in enumerate(results, 1):
            response += f"{i}) t = {t:.2f} —Å ‚Üí T = {T:.3f} —Å ‚Üí g = {g:.2f}\n"

        response += f"\nüìà **–°—Ä–µ–¥–Ω–µ–µ g = {avg_g:.2f} –º/—Å¬≤**\n\n"

        response += "üìä **–ê–Ω–∞–ª–∏–∑ –æ—à–∏–±–æ–∫:**\n"

        # –†–∞–∑–±—Ä–æ—Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        if len(g_values) > 1:
            min_g = min(g_values)
            max_g = max(g_values)
            razbros = max_g - min_g
            response += f"‚Ä¢ –†–∞–∑–±—Ä–æ—Å g: {razbros:.2f} –º/—Å¬≤\n"

            if razbros < 0.5:
                response += "‚úÖ –ò–∑–º–µ—Ä–µ–Ω–∏—è –æ—á–µ–Ω—å —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ!\n"
            elif razbros < 1:
                response += "üëç –•–æ—Ä–æ—à–∞—è –ø–æ–≤—Ç–æ—Ä—è–µ–º–æ—Å—Ç—å\n"
            elif razbros < 2:
                response += "ü§î –°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å ‚Äì —Å—á–∏—Ç–∞–π –∫–æ–ª–µ–±–∞–Ω–∏—è –≤–Ω–∏–º–∞—Ç–µ–ª—å–Ω–µ–µ\n"
            else:
                response += "‚ùå –ë–æ–ª—å—à–æ–π —Ä–∞–∑–±—Ä–æ—Å ‚Äì —á—Ç–æ-—Ç–æ –Ω–µ —Ç–∞–∫\n"

        # –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å
        if error < 5:
            response += "‚úÖ **–û—Ç–ª–∏—á–Ω–æ!** –û—á–µ–Ω—å —Ç–æ—á–Ω—ã–µ –∏–∑–º–µ—Ä–µ–Ω–∏—è\n"
        elif error < 10:
            response += "üëç **–•–æ—Ä–æ—à–æ!** –ù–µ–±–æ–ª—å—à–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å\n"
        elif error < 20:
            response += "ü§î **–°—Ä–µ–¥–Ω–µ.** –ú–æ–∂–µ—Ç, –æ—à–∏–±—Å—è –≤ —Å—á—ë—Ç–µ –∫–æ–ª–µ–±–∞–Ω–∏–π?\n"
        else:
            response += "‚ùå **–ë–æ–ª—å—à–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å.** –ü—Ä–æ–≤–µ—Ä—å:\n"
            response += "‚Ä¢ –¢–æ—á–Ω–æ –ª–∏ 20 –∫–æ–ª–µ–±–∞–Ω–∏–π?\n"
            response += "‚Ä¢ –ü—Ä–∞–≤–∏–ª—å–Ω–æ –ª–∏ –∏–∑–º–µ—Ä–∏–ª –¥–ª–∏–Ω—É?\n"
            response += "‚Ä¢ –ú–∞–ª–µ–Ω—å–∫–∏–π –ª–∏ —É–≥–æ–ª –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è?\n"

        if avg_g > G_TABLE * 1.1:
            response += "‚ö†Ô∏è g > 10.8 ‚Äì –≤–æ–∑–º–æ–∂–Ω–æ, –Ω–∞—Å—á–∏—Ç–∞–ª –º–µ–Ω—å—à–µ 20 –∫–æ–ª–µ–±–∞–Ω–∏–π\n"
        elif avg_g < G_TABLE * 0.9:
            response += "‚ö†Ô∏è g < 8.8 ‚Äì –≤–æ–∑–º–æ–∂–Ω–æ, –Ω–∞—Å—á–∏—Ç–∞–ª –±–æ–ª—å—à–µ 20 –∫–æ–ª–µ–±–∞–Ω–∏–π\n"

        response += f"\nüìä –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: {error:.1f}%"

        await update.message.reply_text(response, parse_mode='Markdown')

    except ValueError:
        await update.message.reply_text('‚ùå –û—à–∏–±–∫–∞! –ü—Ä–∏–º–µ—Ä: 1 40 41 39.5')
    except Exception as e:
        await update.message.reply_text(f'‚ùå –û—à–∏–±–∫–∞: {str(e)}')
```

```python
# @title 12. –†–∞—Å—á—ë—Ç –ø–∞–¥–µ–Ω–∏—è (—Å —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ–º)
async def calculate_fall(update: Update, context: ContextTypes.DEFAULT_TYPE):
    try:
        parts = update.message.text.split()
        if len(parts) < 2:
            await update.message.reply_text('‚ùå –ù—É–∂–Ω–æ –º–∏–Ω–∏–º—É–º 2 —á–∏—Å–ª–∞: h –∏ t1 t2 ...')
            return

        h = float(parts[0].replace(',', '.'))

        # –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –≤—Ä–µ–º–µ–Ω–∞
        times = []
        for val in parts[1:]:
            try:
                t = float(val.replace(',', '.'))
                if t > 0:
                    times.append(t)
                else:
                    await update.message.reply_text(f'‚ö†Ô∏è –í—Ä–µ–º—è {t} <= 0 - –ø—Ä–æ–ø—É—â–µ–Ω–æ')
            except:
                await update.message.reply_text(f'‚ö†Ô∏è "{val}" –Ω–µ —á–∏—Å–ª–æ - –ø—Ä–æ–ø—É—â–µ–Ω–æ')

        if not times:
            await update.message.reply_text('‚ùå –ù–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏!')
            return

        # –ü—Ä–æ–≤–µ—Ä–∫–∏
        if h <= 0:
            await update.message.reply_text('‚ùå –í—ã—Å–æ—Ç–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π')
            return

        if h > 10:
            await update.message.reply_text('‚ö†Ô∏è –í—ã—Å–æ—Ç–∞ –±–æ–ª—å—à–µ 10 –º? –ë—É–¥—å –æ—Å—Ç–æ—Ä–æ–∂–µ–Ω!')

        t_theory = math.sqrt(2 * h / G_TABLE)

        # –†–∞—Å—á—ë—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑–º–µ—Ä–µ–Ω–∏—è
        g_values = []
        results = []

        for t in times:
            if t < 0.1 or t > 3:
                await update.message.reply_text(f'‚ö†Ô∏è –í—Ä–µ–º—è {t} —Å –Ω–µ—Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω–æ - –ø—Ä–æ–ø—É—â–µ–Ω–æ')
                continue

            g = 2 * h / (t**2)
            g_values.append(g)
            results.append((t, g))

        if not g_values:
            await update.message.reply_text('‚ùå –ù–µ—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π –ø–æ—Å–ª–µ –ø—Ä–æ–≤–µ—Ä–æ–∫')
            return

        # –£—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ
        avg_g = sum(g_values) / len(g_values)
        error = abs(avg_g - G_TABLE) / G_TABLE * 100

        # –ê–Ω–∞–ª–∏–∑
        response = "üìä **–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ –ø–∞–¥–µ–Ω–∏—è**\n\n"
        response += f"üìè h = {h} –º\n"
        response += f"üìê –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ t = {t_theory:.3f} —Å\n"
        response += f"üìä –ò–∑–º–µ—Ä–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏: {len(times)}\n\n"

        response += "**–û—Ç–¥–µ–ª—å–Ω—ã–µ –∏–∑–º–µ—Ä–µ–Ω–∏—è:**\n"
        for i, (t, g) in enumerate(results, 1):
            dev = abs(t - t_theory) / t_theory * 100
            response += f"{i}) t = {t:.3f} —Å ‚Üí g = {g:.2f}"
            if dev < 10:
                response += " ‚úì\n"
            else:
                response += f" (–æ—Ç–∫–ª. {dev:.0f}%)\n"

        response += f"\nüìà **–°—Ä–µ–¥–Ω–µ–µ g = {avg_g:.2f} –º/—Å¬≤**\n\n"

        response += "üìä **–ê–Ω–∞–ª–∏–∑ –æ—à–∏–±–æ–∫:**\n"

        # –†–∞–∑–±—Ä–æ—Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
        if len(g_values) > 1:
            min_g = min(g_values)
            max_g = max(g_values)
            razbros = max_g - min_g
            response += f"‚Ä¢ –†–∞–∑–±—Ä–æ—Å g: {razbros:.2f} –º/—Å¬≤\n"

            if razbros < 1:
                response += "‚úÖ –ò–∑–º–µ—Ä–µ–Ω–∏—è —Å—Ç–∞–±–∏–ª—å–Ω—ã–µ\n"
            elif razbros < 2:
                response += "üëç –°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–±—Ä–æ—Å\n"
            else:
                response += "‚ùå –ë–æ–ª—å—à–æ–π —Ä–∞–∑–±—Ä–æ—Å ‚Äì —Ç—Ä–µ–Ω–∏—Ä—É–π —Ä–µ–∞–∫—Ü–∏—é\n"

        # –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å
        if error < 5:
            response += "‚úÖ **–û—Ç–ª–∏—á–Ω–æ!** –¢—ã –æ—Ç–ª–∏—á–Ω–æ –ø–æ–π–º–∞–ª –º–æ–º–µ–Ω—Ç!\n"
        elif error < 10:
            response += "üëç **–•–æ—Ä–æ—à–æ!** –î–ª—è —Ä—É—á–Ω–æ–≥–æ —Å–µ–∫—É–Ω–¥–æ–º–µ—Ä–∞ –æ—Ç–ª–∏—á–Ω–æ\n"
        elif error < 20:
            response += "ü§î **–°—Ä–µ–¥–Ω–µ.** –ü–æ–ø—Ä–æ–±—É–π —Å –ø–æ–º–æ—â–Ω–∏–∫–æ–º\n"
        else:
            response += "‚ùå **–ë–æ–ª—å—à–∞—è –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å.** –ü—Ä–æ–±–ª–µ–º—ã:\n"
            response += "‚Ä¢ –°–ª–∏—à–∫–æ–º –ø–æ–∑–¥–Ω–æ –∑–∞–ø—É—Å—Ç–∏–ª —Å–µ–∫—É–Ω–¥–æ–º–µ—Ä\n"
            response += "‚Ä¢ –°–ª–∏—à–∫–æ–º —Ä–∞–Ω–æ –æ—Å—Ç–∞–Ω–æ–≤–∏–ª\n"
            response += "‚Ä¢ –ü–æ–ø—Ä–æ—Å–∏ –ø–æ–º–æ—â–Ω–∏–∫–∞\n"

        # –ê–Ω–∞–ª–∏–∑ –≤—Ä–µ–º–µ–Ω–∏
        avg_t = sum(times) / len(times)
        if avg_t > t_theory * 1.2:
            response += "‚ö†Ô∏è –í—Ä–µ–º—è –±–æ–ª—å—à–µ —Ç–µ–æ—Ä–∏–∏ ‚Äì –ø–æ–∑–¥–Ω–æ –∑–∞–ø—É—Å—Ç–∏–ª?\n"
        elif avg_t < t_theory * 0.8:
            response += "‚ö†Ô∏è –í—Ä–µ–º—è –º–µ–Ω—å—à–µ —Ç–µ–æ—Ä–∏–∏ ‚Äì –∑–∞–ø—É—Å—Ç–∏–ª –¥–æ –±—Ä–æ—Å–∫–∞?\n"
        else:
            response += "‚úÖ –í—Ä–µ–º—è –±–ª–∏–∑–∫–æ –∫ —Ç–µ–æ—Ä–∏–∏\n"

        response += f"\nüìä –ü–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: {error:.1f}%"

        await update.message.reply_text(response, parse_mode='Markdown')

    except ValueError:
        await update.message.reply_text('‚ùå –û—à–∏–±–∫–∞! –ü—Ä–∏–º–µ—Ä: 2 0.64 0.63 0.65')
    except Exception as e:
        await update.message.reply_text(f'‚ùå –û—à–∏–±–∫–∞: {str(e)}')
```

```python
# @title 13. –û–±—Ä–∞–±–æ—Ç–∫–∞ —Å–æ–æ–±—â–µ–Ω–∏–π
async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    text = update.message.text

    # –ö–Ω–æ–ø–∫–∏ –º–µ–Ω—é
    if text == 'üîç –§–æ–∫—É—Å –ª–∏–Ω–∑—ã':
        await lens_experiment(update, context)
    elif text == '‚è± –ú–∞—è—Ç–Ω–∏–∫':
        await pendulum_experiment(update, context)
    elif text == 'üìä –°–≤–æ–±–æ–¥–Ω–æ–µ –ø–∞–¥–µ–Ω–∏–µ':
        await fall_experiment(update, context)
    elif text == '‚ùì –ü–æ–º–æ—â—å':
        await help_command(update, context)
    else:
        # –ï—Å–ª–∏ –Ω–µ –∫–Ω–æ–ø–∫–∞, –∑–Ω–∞—á–∏—Ç —ç—Ç–æ –¥–∞–Ω–Ω—ã–µ
        mode = context.user_data.get('mode')

        if mode == 'lens':
            await calculate_lens(update, context)
        elif mode == 'pendulum':
            await calculate_pendulum(update, context)
        elif mode == 'fall':
            await calculate_fall(update, context)
        else:
            await update.message.reply_text('–°–Ω–∞—á–∞–ª–∞ –≤—ã–±–µ—Ä–∏ –æ–ø—ã—Ç –≤ –º–µ–Ω—é!')
```

```python
# @title 14. –ó–ê–ü–£–°–ö (–í–°–¢–ê–í–¨ –¢–û–ö–ï–ù)
TOKEN = "–í–°–¢–ê–í–¨_–°–Æ–î–ê_–°–í–û–ô_–¢–û–ö–ï–ù"

def main():
    app = Application.builder().token(TOKEN).build()

    app.add_handler(CommandHandler("start", start))
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))

    print("üöÄ –ë–æ—Ç –∑–∞–ø—É—â–µ–Ω! –ò–¥–∏ –≤ Telegram –∏ –Ω–∞–∂–º–∏ /start")
    print("‚úÖ –ë–æ—Ç –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å, –ø–æ–∫–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —ç—Ç–∞ —è—á–µ–π–∫–∞")
    app.run_polling()

if __name__ == '__main__':
    if TOKEN == "–í–°–¢–ê–í–¨_–°–Æ–î–ê_–°–í–û–ô_–¢–û–ö–ï–ù":
        print("‚ùå –í—Å—Ç–∞–≤—å —Ç–æ–∫–µ–Ω –≤ —è—á–µ–π–∫—É 14!")
    else:
        main()
```
